Create the charactersFixed table.

````python
%%sql

create table s23_w4111_hw2_xf2198.charactersFixed (
    characterId VARCHAR(500) NOT NULL,
    characterName VARCHAR(500) NOT NULL,
    characterImdbID VARCHAR(500) NOT NULL,
    characterLink VARCHAR(500) NOT NULL,
    actorNConst VARCHAR(500) NOT NULL,
    actorLink VARCHAR(500) NOT NULL,
    characterImageFull VARCHAR(500) NOT NULL,
    characterImageThumb VARCHAR(500) NOT NULL,
    kingsguard VARCHAR(500) NOT NULL,
    royal VARCHAR(500) NOT NULL,
    PRIMARY KEY (characterId));
````

Generating the characterID.

````python
%%sql

create trigger s23_w4111_hw2_xf2198.charactersFixed_before_insert
before insert on s23_w4111_hw2_xf2198.charactersFixed
for each row
begin 
declare prefix VARCHAR(500);
set prefix 
= SUBSTRING_INDEX(NEW.characterName, ' ', 1);
set NEW.characterId 
= CONCAT(prefix, (select COUNT(*)+1 
from s23_w4111_hw2_xf2198.charactersFixed));
end;
````

````python
%%sql

insert into s23_w4111_hw2_xf2198.charactersFixed 
(characterName, 
 characterImdbID, 
 characterLink, 
 actorNConst, 
 actorLink,
 characterImageFull, 
 characterImageThumb, 
 kingsguard, 
 royal)
select characterName, 
IFNULL(replace(characterLink, '/character/', ''), 'NaN'),  
IFNULL(characterLink, 'NaN'),
IFNULL(actorLink, 'NaN'), 
IFNULL(CONCAT('/names/', actorLink), 'NaN'),
IFNULL(characterImageFull, 'NaN'), 
IFNULL(characterImageThumb, 'NaN'),
IFNULL(kingsguard, 'NaN'),
IFNULL(royal, 'NaN')
from s23_w4111_hw2_xf2198.characters;
````

````python
%%sql

create table s23_w4111_hw2_xf2198.name_professions (
  name VARCHAR(100) NOT NULL,
  profession_id INT NOT NULL,
  PRIMARY KEY (name, profession_id),
  FOREIGN KEY (profession_id) references primary_professions(profession_id),
  FOREIGN KEY (name) references name_basics_all_fixed(primaryName)
);
````

````python
%%sql

insert into s23_w4111_hw2_xf2198.name_professions (name, profession_id)
select s23_w4111_hw2_xf2198.name_basics_all.primaryName, 
s23_w4111_hw2_xf2198.primary_professions.profession_id
from s23_w4111_hw2_xf2198.name_basics_all
inner join (
  select s23_w4111_hw2_xf2198.name_basics_all.nconst, SUBSTRING_INDEX(SUBSTRING_INDEX(s23_w4111_hw2_xf2198.name_basics_all.primaryProfession, ',', numbers.n), ',', -1) as profession
  from s23_w4111_hw2_xf2198.name_basics_all
  inner join (
    select 0 as n UNION ALL SELECT 1 UNION ALL SELECT 2 UNION ALL SELECT 3 UNION ALL SELECT 4 UNION ALL SELECT 5 UNION ALL SELECT 6 UNION ALL SELECT 7 UNION ALL SELECT 8 UNION ALL SELECT 9
  ) as numbers on CHAR_LENGTH(s23_w4111_hw2_xf2198.name_basics_all.primaryProfession) - CHAR_LENGTH(REPLACE(s23_w4111_hw2_xf2198.name_basics_all.primaryProfession, ',', '')) >= numbers.n - 1
) as subquery2 on s23_w4111_hw2_xf2198.name_basics_all.nconst = subquery2.nconst
inner join s23_w4111_hw2_xf2198.primary_professions on subquery2.profession = s23_w4111_hw2_xf2198.primary_professions.profession_name;
````
